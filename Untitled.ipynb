# Gillespie Model

```python
import random
import math

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
# Maximum elapsed time of simulation
T = np.float(10000.0)

# Start time
t = np.float(0.0)

# Spatial parameter
V = 100.0

# Rate of triangle binding with enzyme 
triangle_rate = 8

# Rate of circle binding with enzyme 
circle_rate = 0.9

number_of_enzyme = 10
number_of_triangle = 100
number_of_circle = 10

# initialize results list
SIR_data = []
SIR_data.append((t, number_of_triangle, number_of_circle))

# main loop
while t < T:

    if number_of_enzyme == 0:
        break

    w1 = triangle_rate * number_of_triangle / V
    w2 = circle_rate * number_of_circle / V
    W = w1 + w2
    
    if W == 0:
        break
        
    dt = -np.log(np.random.random_sample()) / W
    t = t + dt

    if np.random.random_sample() < w1 / W:
        number_of_triangle = number_of_triangle - 1

    else:
        number_of_circle = number_of_circle - 1

    SIR_data.append((t, number_of_triangle, number_of_circle))
    
data = [list(t) for t in zip(*SIR_data)]
plt.plot(data[0], data[1])
plt.plot(data[0], data[2])
plt.title("Population vs Time")
plt.legend(["Triangle", "Circle"])
plt.xlabel("Time (Step)")
plt.ylabel("Population Count")
plt.show()
```
